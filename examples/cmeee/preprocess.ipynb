```python
import json
```

```python
for filepath in ['/content/drive/MyDrive/datasets/CMeEE/CMeEE_train.json','/content/drive/MyDrive/datasets/CMeEE/CMeEE_dev.json' ]:
    with open(filepath, encoding='utf8')as f:
        for data in json.load(f):
            has_error = False
            for entity in data.get('entities', []):
                if entity['entity'] != data['text'][entity['start_idx']: entity['end_idx']]:
                    guest_start = data['text'].index(entity['entity'])
                    print(f'filepath: {filepath.split("/")[-1]}, text: {data["text"]}, guest: {guest_start}-{guest_start+len(entity["entity"])}, error: {entity["start_idx"]}-{entity["end_idx"]}')
                    has_error = True
            if has_error is True:
                break        
```

```python
data_list = []
for filepath in ['/content/drive/MyDrive/datasets/CMeEE/CMeEE_train.json','/content/drive/MyDrive/datasets/CMeEE/CMeEE_dev.json' ]:
    with open(filepath, encoding='utf8')as f:
        #data_list.extend(json.load(f))
        for data in json.load(f):
            for entity in data.get('entities', []):
                entity['end_idx'] += 1
            data_list.append(data)
```

```python
len(data_list), data_list[0]
```

```python
with open('/content/drive/MyDrive/datasets/CMeEE/data.json', 'w', encoding='utf8')as f:
    json.dump(data_list, f, ensure_ascii=False, indent=2)
```
